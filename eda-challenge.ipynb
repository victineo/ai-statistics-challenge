```python
# Importing pandas
import pandas as pd
```

```python
# Importing matplotlib
import matplotlib.pyplot as plt
```

```python
# Billing dictionary
dict_sales = {
    'data_ref': ['2023-01-01', '2020-02-01', '2021-03-01', '2022-04-01', '2023-05-01', '2023-06-01', '2020-07-01', '2021-08-01', '2022-09-01', '2023-10-01','2022-11-01', '2023-12-01'],
    'value': [400000, 890000, 760000, 430000, 920000,340000, 800000, 500000, 200000, 900000,570000, 995000]
}
```

```python
dict_sales
```

```python
# Transforming the dictionary into a DataFrame
df_sales = pd.DataFrame.from_dict(dict_sales)
```

```python
df_sales
```

```python
# Showing the mean sales
df_sales['value'].mean()
```

```python
# Creating a vertical bar chart showing the reference month and value
df_sales.plot.bar(x='data_ref', y='value')
```

```python
# Creating a line chart showing the reference month and value
df_sales.plot.line(x='data_ref', y='value')
```
